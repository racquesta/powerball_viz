```python
from states import states
from random import randint
import json
```

```python
states_list = list(states)
```

```python
final_geojson = {"type": "FeatureCollection","features":[]}
for state in states_list:
    data = json.load(open("us_territories_geojson/{}/shape.geojson".format(state)))
    state_name = states[state]
    dict_add = {"type":"Feature",
                "properties":{"state_name":"{}".format(state_name), "info_1":"{}".format(randint(10, 100)), "info_2": "{}".format(randint(500000, 5000))},
                "geometry": data
               }
    final_geojson['features'].append(dict_add)
```

```python
with open('us_territories.geojson', 'w') as f:
    json.dump(final_geojson, f)
```
